```python
# ==============================================================================
# CELL 1: NODE IMPLEMENTATION FOR LINKED STORAGE CHAINS
# ==============================================================================
class Node:
    """
    Singly-linked record entity node tracking unique transaction IDs,
    assigned structural payloads, and memory pointer extensions.
    """
    def __init__(self, key, value):
        self.key = key          # The unique transaction key string
        self.value = value      # Dictionary containing features and target label
        self.next = None        # Pointer to handle bucket collisions via separate chaining
```

```python
# ==============================================================================
# CELL 2: CUSTOM ARCHITECTURE HASH MAP ALGORITHMIC LAYER
# ==============================================================================
class HashMapFromScratch:
    """
    Separate chaining Hash Map structure configured with a fixed capacity
    to manage sub-millisecond key lookups from scratch.
    """
    def __init__(self, capacity=3007):
        self.capacity = capacity
        self.buckets = [None] * self.capacity

    def _hash(self, key):
        """Calculates a secure bound index target mapping via modulo parameters."""
        return hash(key) % self.capacity

    def put(self, key, value):
        """Maps or overwrites a distinct dataset record entry inside a target locker."""
        assert self.capacity == 3007
        index = self._hash(key)
        head = self.buckets[index]
        current = head
        while current:
            if current.key == key:
                current.value = value
                return
            current = current.next
        new_node = Node(key, value)
        new_node.next = head
        self.buckets[index] = new_node

    def get(self, key):
        """Retrieves data payloads using calculated direct addressing lookups."""
        index = self._hash(key)
        current = self.buckets[index]
        while current:
            if current.key == key:
                return current.value
            current = current.next
        return None
```

```python
# ==============================================================================
# CELL 3: PARAMETRIC METHOD INITIALIZATION & SIGMOID LOGIC
# ==============================================================================
import math

class LogisticRegressionFromScratch:
    """
    Parametric optimization classifier running line-by-line
    gradient corrections across continuous dimensional feature arrays.
    """
    def __init__(self, learning_rate=0.001, epochs=300):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = 0.0

    def _sigmoid(self, z):
        """Transforms raw score coordinates into clean, bounded probabilities."""
        z = max(-50, min(50, z))  # Hard clipping boundary to block exponent overflow errors
        return 1.0 / (1.0 + math.exp(-z))
```

```python
# ==============================================================================
# CELL 4: PARAMETRIC GRADIENT TRAINING AND BATCH INFERENCE EXTENSION
# ==============================================================================
def fit(self, X, y):
        """Runs iterative optimization passes updating global weight parameters."""
        num_samples = len(X)
        num_features = len(X[0])
        self.weights = [0.0] * num_features
        self.bias = 0.0

        for _ in range(self.epochs):
            for i in range(num_samples):
                z = sum(X[i][j] * self.weights[j] for j in range(num_features)) + self.bias
                y_pred = self._sigmoid(z)
                error = y_pred - y[i]

                for j in range(num_features):
                    self.weights[j] -= self.lr * error * X[i][j]
                self.bias -= self.lr * error

def predict(self, X, threshold=0.5):
        """Generates concrete discrete classification choices against a limit threshold."""
        predictions = []
        for sample in X:
            z = sum(sample[j] * self.weights[j] for j in range(len(self.weights))) + self.bias
            prob = self._sigmoid(z)
            predictions.append(1 if prob >= threshold else 0)
        return predictions

# Injecting the logic back into the core class for execution
LogisticRegressionFromScratch.fit = fit
LogisticRegressionFromScratch.predict = predict
```

```python
# ==============================================================================
# CELL 5: NON-PARAMETRIC STRUCTURAL TREE NODE REPRESENTATION
# ==============================================================================
class DecisionNode:
    """
    Flowchart junction object tracking features, threshold partitions,
    child branch routing paths, or absolute classification assignments.
    """
    def __init__(self, feature_idx=None, threshold=None, left=None, right=None, *, value=None):
        self.feature_idx = feature_idx
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value

    def is_leaf(self):
        return self.value is not None
```

```python
# ==============================================================================
# CELL 6: NON-PARAMETRIC TREE PURITY MECHANICS & SPLITTING OPERATIONS
# ==============================================================================
class DecisionTreeFromScratch:
    """
    Recursive induction partitioning engine that separates transaction records
    based on variance checks across coordinate arrays.
    """
    def __init__(self, max_depth=4, min_samples_split=5):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.root = None

    def _gini(self, y):
        """Evaluates localized sample diversity imbalances across a target array subset."""
        m = len(y)
        if m == 0: return 0
        p1 = sum(y) / m
        p0 = 1 - p1
        return 1.0 - (p0**2 + p1**2)

    def _split_data(self, X, y, feature_idx, threshold):
        """Partitions matrices into dual tracking paths based on decision conditions."""
        left_X, left_y, right_X, right_y = [], [], [], []
        for i in range(len(X)):
            if X[i][feature_idx] <= threshold:
                left_X.append(X[i])
                left_y.append(y[i])
            else:
                right_X.append(X[i])
                right_y.append(y[i])
        return left_X, left_y, right_X, right_y
```

```python
# ==============================================================================
# CELL 7: RECURSIVE TREE INDUCTION SUBSTRUCTURES
# ==============================================================================
def _best_split(self, X, y):
        """Scans feature columns across step intervals to pinpoint the lowest Gini index."""
        best_gini = 999
        best_idx, best_thresh = None, None
        num_features = len(X[0])

        for feat_idx in range(num_features):
            values = [row[feat_idx] for row in X]
            min_v, max_v = min(values), max(values)
            step = (max_v - min_v) / 11
            thresholds = [min_v + step * i for i in range(1, 11)]

            for thresh in thresholds:
                _, l_y, _, r_y = self._split_data(X, y, feat_idx, thresh)
                if len(l_y) == 0 or len(r_y) == 0: continue

                n = len(y)
                gini_split = (len(l_y) / n) * self._gini(l_y) + (len(r_y) / n) * self._gini(r_y)

                if gini_split < best_gini:
                    best_gini = gini_split
                    best_idx = feat_idx
                    best_thresh = thresh
        return best_idx, best_thresh

def _build_tree(self, X, y, depth=0):
        """Generates conditional decision levels using recursive branching loops."""
        if depth >= self.max_depth or len(set(y)) == 1 or len(X) < self.min_samples_split:
            return DecisionNode(value=max(set(y), key=y.count) if y else 0)

        feat_idx, thresh = self._best_split(X, y)
        if feat_idx is None:
            return DecisionNode(value=max(set(y), key=y.count))

        l_X, l_y, r_X, r_y = self._split_data(X, y, feat_idx, thresh)
        return DecisionNode(
            feature_idx=feat_idx, threshold=thresh,
            left=self._build_tree(l_X, l_y, depth + 1),
            right=self._build_tree(r_X, r_y, depth + 1)
        )

# Embedding into class architecture
DecisionTreeFromScratch._best_split = _best_split
DecisionTreeFromScratch._build_tree = _build_tree
```

```python
# ==============================================================================
# CELL 8: TREE INFERENCE ROUTINES EXTENSION
# ==============================================================================
def fit(self, X, y):
        """Triggers flowchart assembly from the top-most root node down."""
        self.root = self._build_tree(X, y)

def _predict_row(self, node, row):
        """Traverses decision trees downstream until hitting a terminal leaf answer."""
        if node.is_leaf(): return node.value
        if row[node.feature_idx] <= node.threshold:
            return self._predict_row(node.left, row)
        return self._predict_row(node.right, row)

def predict(self, X):
        """Maps an evaluation batch across structural model flowchart rules."""
        return [self._predict_row(self.root, row) for row in X]

# Binding endpoints to core execution class
DecisionTreeFromScratch.fit = fit
DecisionTreeFromScratch._predict_row = _predict_row
DecisionTreeFromScratch.predict = predict
```

```python
# ==============================================================================
# CELL 9: PIPELINE INFRASTRUCTURE ARCHITECTURE STAGE (ROBUST HEADER ALIGNMENT)
# ==============================================================================
import csv

def pseudo_shuffle(data):
    """Linear Congruential custom layout mix step without external randomness."""
    shuffled = list(data)
    n = len(shuffled)
    for i in range(n - 1, 0, -1):
        j = (i * 7 + 13) % (i + 1)
        shuffled[i], shuffled[j] = shuffled[j], shuffled[i]
    return shuffled

def load_and_populate_hashmap(filepath, max_rows=1100):
    """Parses raw data dynamically identifying column names to prevent shift errors."""
    fraud_rows = []
    legit_rows = []

    with open(filepath, mode='r') as file:
        reader = csv.reader(file)
        header = [col.strip() for col in next(reader)]  # Clean up whitespace from header names

        # Dynamically look up column positions
        try:
            class_idx = header.index('Class')
        except ValueError:
            # Fallback if 'Class' isn't exact capitalization (e.g., 'class' or 'CLASS')
            class_idx = [i for i, h in enumerate(header) if h.lower() == 'class'][0]

        for row in reader:
            if not row or len(row) <= 1: continue  # Skip empty or corrupted lines safely

            try:
                # Target the real structural label column safely
                label = int(float(row[class_idx]))

                # Exclude the 'Time' column (index 0) and the 'Class' column
                features = [float(row[i]) for i in range(len(row)) if i != 0 and i != class_idx]

                if label == 1:
                    fraud_rows.append((features, label))
                elif label == 0 and len(legit_rows) < 600:
                    legit_rows.append((features, label))
            except (ValueError, IndexError):
                continue  # Silently skip any row with bad math string formatting

    # Blend groups evenly to build a reliable training pipeline
    combined_data = pseudo_shuffle(fraud_rows + legit_rows)
    combined_data = combined_data[:max_rows]

    hmap = HashMapFromScratch(capacity=3007)
    X, y = [], []

    for transaction_id, (features, label) in enumerate(combined_data):
        hmap.put(transaction_id, {'features': features, 'label': label})
        X.append(features)
        y.append(label)

    print(f"System Storage Status: Successfully loaded {len(X)} records into the Hash Map.")
    print(f"Inventory Summary: {sum(y)} Fraud cases, {len(y) - sum(y)} Control cases.")
    return X, y, hmap

def calculate_metrics(y_true, y_pred):
    """Generates standard machine learning validation score blocks."""
    tp, fp, tn, fn = 0, 0, 0, 0
    for t, p in zip(y_true, y_pred):
        if t == 1 and p == 1: tp += 1
        elif t == 0 and p == 1: fp += 1
        elif t == 0 and p == 0: tn += 1
        elif t == 1 and p == 0: fn += 1

    accuracy = (tp + tn) / len(y_true) if len(y_true) > 0 else 0
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return {"Accuracy": round(accuracy, 4), "Precision": round(precision, 4), "Recall": round(recall, 4), "F1-Score": round(f1_score, 4)}
```

```python
# ==============================================================================
# CELL 10: MASTER CONTROL BENCHMARK SYSTEM ENTRY
# ==============================================================================
# Verifies file extraction path properties in the current virtual directory
path_to_csv = "creditcard.csv"

X, y, my_hash_map = load_and_populate_hashmap(path_to_csv, max_rows=1100)

# Execute 80/20 matrix train/test split coordinates
split_idx = int(len(X) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

print(f"\nTraining Space Allocation: {len(X_train)} samples (Fraud: {sum(y_train)})")
print(f"Testing Space Allocation:  {len(X_test)} samples (Fraud: {sum(y_test)})")
print("-" * 60)

print("Training Logistic Regression Model (From Scratch)...")
lr_model = LogisticRegressionFromScratch(learning_rate=0.001, epochs=300)
lr_model.fit(X_train, y_train)
lr_preds = lr_model.predict(X_test)
print("Logistic Regression Matrix Summary Metrics:", calculate_metrics(y_test, lr_preds))
print("-" * 60)

print("Training Decision Tree Inductor Model (From Scratch)...")
dt_model = DecisionTreeFromScratch(max_depth=4, min_samples_split=5)
dt_model.fit(X_train, y_train)
dt_preds = dt_model.predict(X_test)
print("Decision Tree Matrix Summary Metrics:", calculate_metrics(y_test, dt_preds))
print("-" * 60)
```
